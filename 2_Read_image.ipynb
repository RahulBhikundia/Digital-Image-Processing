```python
import cv2
from matplotlib import pyplot as plt

img = cv2.imread("Hut.png", 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis('off')  # Turn off axis
plt.show()
```

```python
print(img.shape)
print(img.size)
```
