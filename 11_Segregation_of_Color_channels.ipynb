```python
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("Hut.jpg")
```

```python
# Splitting Image into RED, BLUE, GREEN Channel
blue, green, red = cv2.split(image)
```

```python
# Display each image seperately
plt.subplot(141), plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB)), plt.title("Original"), plt.axis('off')
plt.subplot(142), plt.imshow(cv2.cvtColor(blue,cv2.COLOR_BGR2RGB)), plt.title("Blue"), plt.axis('off')
plt.subplot(143), plt.imshow(cv2.cvtColor(green,cv2.COLOR_BGR2RGB)), plt.title("Green"), plt.axis('off')
plt.subplot(144), plt.imshow(cv2.cvtColor(red,cv2.COLOR_BGR2RGB)), plt.title("Red"), plt.axis('off')
plt.show()
```
