```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("Hut.jpg",cv2.IMREAD_GRAYSCALE)
```

```python
hist = cv2.calcHist([img], [0], None, [256], [0,256])
plt.subplot(234), plt.plot(hist)
plt.xlabel('Bins')
plt.ylabel('# of pixels')
plt.title('Histograms')

eq = cv2.equalizeHist(img)
hist = cv2.calcHist([eq], [0], None, [256], [0,256])
plt.subplot(236), plt.plot(hist)
plt.xlabel('Bins')
plt.ylabel('# of pixels')
plt.title('Equalized Histogram')

plt.subplot(231), plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("Original")
plt.subplot(233), plt.imshow(cv2.cvtColor(eq,cv2.COLOR_BGR2RGB))
plt.title("Equalized")
plt.show()
```
