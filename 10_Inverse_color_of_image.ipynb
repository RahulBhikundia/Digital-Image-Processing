```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("Hut.jpg")
invert_image = 255 - image
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(121), plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB)), plt.axis('off')
plt.subplot(122), plt.imshow(cv2.cvtColor(invert_image,cv2.COLOR_BGR2RGB)), plt.axis('off')
plt.show()
```
