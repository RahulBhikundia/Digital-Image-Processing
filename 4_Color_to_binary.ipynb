```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("hut.jpg",1)
```

```python
if img is None:
    print("Error: Unable to load the image")
else:
    image_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    (thrash,binary_image) = cv2.threshold(gray, 175, 255, cv2.THRESH_BINARY)

    plt.subplot(131), plt.imshow(image_rgb), plt.title("Original Image"), plt.axis('off')
    plt.subplot(132), plt.imshow(gray, cmap='gray'), plt.title("Grayscale Image"), plt.axis('off')
    plt.subplot(133), plt.imshow(binary_image, cmap='gray'), plt.title("Binary Image"), plt.axis('off')
    plt.show()
```

```python
cv2.imwrite("gray.jpg",gray)
```
