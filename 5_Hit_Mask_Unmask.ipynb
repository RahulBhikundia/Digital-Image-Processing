```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread("hut.jpg",0)
image = cv2.imread("hut.jpg")
```

```python
hist = cv2.calcHist([img],[0],None,[256],[0,256])
histB = cv2.calcHist([image],[0],None,[256],[0,256])
histG = cv2.calcHist([image],[1],None,[256],[0,256])
histR = cv2.calcHist([image],[2],None,[256],[0,256])
```

```python
mask = np.zeros(img.shape[:2], np.uint8)
mask[100:300,100:400] = 255
```

```python
masked_image = cv2.bitwise_and(img, img, mask=mask)
hist_mask = cv2.calcHist([img],[0],mask,[256],[0,256])
```

```python
plt.subplot(231), plt.imshow(img, cmap='gray'), plt.title("Original")
plt.subplot(232), plt.imshow(mask, cmap='gray'), plt.title("Mask")
plt.subplot(233), plt.imshow(masked_image, cmap='gray'), plt.title("Masked Image")
plt.subplot(234), plt.plot(hist), plt.title("Without Mask")
plt.subplot(236), plt.plot(hist_mask), plt.title("With Mask")
plt.show()
```

```python
plt.title("Histogram of all RGB colors")
plt.plot(histR, color="Red")
plt.plot(histG, color="Green")
plt.plot(histB, color="Blue")
plt.show()
```
